# Adaptive Per-Head KV Cache Compression Demo

This notebook demonstrates how to use the `AdaptivePerHeadPress` to apply learned per-layer per-head compression strategies to KV cache compression.

## Overview

Instead of using a fixed compression ratio for all layers and heads, `AdaptivePerHeadPress` uses a trained policy network to dynamically allocate different compression budgets to each layer and attention head based on their KV cache statistics.

### Prerequisites

1. Train a policy network using the code in `policy/` directory
2. Save the checkpoint (e.g., `ckpts/policy/policy_epoch10.pt`)
3. Load the checkpoint in this wrapper

### Key Features

- **Adaptive**: Different heads get different compression ratios based on learned importance
- **Flexible**: Works with any `ScorerPress`-based compression method
- **Efficient**: Policy network runs only once during prefill, minimal overhead
